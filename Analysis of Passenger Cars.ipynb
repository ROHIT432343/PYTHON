```python
import pandas as pd
df=pd.read_csv('DS1_C4_S7_Car_Data_Challenge.csv')
```

```python
# Assuming your DataFrame is named 'df'
columns_to_retain = ['Cylinders', 'Kerb_Weight', 'Make', 'Displacement', 'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 'Highway_Mileage_km_litre','Power', 'Torque']

# Retain only the specified columns
df = df[columns_to_retain]

# Drop rows with missing values in the specified columns
drow=df.dropna(subset=['Power', 'Torque'], inplace=True)
```

```python
# Retain only the required columns
selected_columns = ['Cylinders', 'Kerb_Weight', 'Make', 'Displacement', 'Fuel_Tank_Capacity_litre',
                    'City_Mileage_km_litre', 'Highway_Mileage_km_litre', 'Power', 'Torque']

df_selected = df[selected_columns]

# Drop rows with missing values in the specified columns
df_selected_cleaned = df_selected.dropna(subset=['Cylinders', 'Kerb_Weight', 'Make', 'Displacement',
                                                 'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre',
                                                 'Highway_Mileage_km_litre', 'Power', 'Torque'])

# Display the cleaned DataFrame
print(df_selected_cleaned)
```

###### task 2

```python
# Cleaning Kerb_Weight column
df_selected_cleaned['Kerb_Weight'] = df_selected_cleaned['Kerb_Weight'].str.replace(r"kg", '')  # Remove 'kg'
df_selected_cleaned['Kerb_Weight'] = pd.to_numeric(df_selected_cleaned['Kerb_Weight'], errors='coerce')  # Convert to numeric

# Cleaning Power column
df_selected_cleaned['Power'] = df_selected_cleaned['Power'].str.strip()  # Stripping white spaces if any
df_selected_cleaned['Power'] = df_selected_cleaned['Power'].str.extract(r'(\d+)').astype(float)  # Extracting the numeric part

# Convert Kerb_Weight from kg to pounds
df_selected_cleaned['Kerb_Weight'] = df_selected_cleaned['Kerb_Weight'] * 2.2

# Display the cleaned DataFrame
print(df_selected_cleaned)
```

##### task 3

```python


# Cleaning Torque column
df_selected_cleaned['Torque'] = df_selected_cleaned['Torque'].str.strip()  # Stripping white spaces if any
df_selected_cleaned['Torque'] = df_selected_cleaned['Torque'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*", '', regex=True)  # Extracting the numeric part
df_selected_cleaned['Torque'] = pd.to_numeric(df_selected_cleaned['Torque'], errors='coerce')  # Convert to numeric

# Display the cleaned DataFrame
print(df_selected_cleaned)
```

###### task 3

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate and display mean, median, standard deviation, and coefficient of variation for each manufacturer
stats_columns = ['Displacement', 'Kerb_Weight', 'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 'Highway_Mileage_km_litre', 'Power', 'Torque']

for col in stats_columns:
    manufacturer_stats = df_selected_cleaned.groupby('Make')[col].agg(['mean', 'median', 'std'])
    manufacturer_stats['coefficient_of_variation'] = manufacturer_stats['std'] / manufacturer_stats['mean']
    print(f"\nStatistics for {col}:\n")
    print(manufacturer_stats)
```

```python
# Display box plot for the 'Displacement' column for each manufacturer
plt.figure(figsize=(12, 6))
sns.boxplot(x='Make', y='Displacement', data=df_selected_cleaned)
plt.title('Displacement Distribution for Each Manufacturer')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Selecting data for the specified manufacturers
selected_manufacturers = ['Renault', 'Toyota', 'Honda']
df_selected_manufacturers = df_selected_cleaned[df_selected_cleaned['Make'].isin(selected_manufacturers)]

# List of features to analyze
features_to_analyze = ['Displacement', 'Kerb_Weight', 'Power', 'Torque', 'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 'Highway_Mileage_km_litre']

# Correlation matrix
correlation_matrix = df_selected_manufacturers[features_to_analyze].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features for Selected Manufacturers')
plt.show()

# Visualize the correlation of features with City Mileage and Highway Mileage
for mileage_type in ['City_Mileage_km_litre', 'Highway_Mileage_km_litre']:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=correlation_matrix[mileage_type].drop(index=[mileage_type]).index,
                y=correlation_matrix[mileage_type].drop(index=[mileage_type]).values)
    plt.title(f'Correlation of Features with {mileage_type}')
    plt.xlabel('Features')
    plt.ylabel('Correlation Coefficient')
    plt.show()
```
