```python
inventory = [
    {"item": "biscuits", "quantity": 5, "cost_per_item": 2.50},
    {"item": "cereals", "quantity": 10, "cost_per_item": 3.50},
    {"item": "chicken", "quantity": 20, "cost_per_item": 13.30},
    {"item": "oats", "quantity": 20, "cost_per_item": 3.90},
    {"item": "rice", "quantity": 30, "cost_per_item": 4.50},]
# Delivery charge based on distance
def calculate_delivery_charge(distance):
    if distance <= 5:
        return 10
    elif distance <= 10:
        return 12
    elif distance <= 15:
        return 14
    else:
        return 20
customer_cart = []
total_cost = 0
while True:
    print("S.No   Items      Quantity   Cost/Item")
    for i, item in enumerate(inventory, start=1):
         print(f"{i}\t{item['item']}\t{item['quantity']}\t${item['cost_per_item']:.2f}")

    customer_name = input("Enter your name: ")
    delivery_address = input("Enter your delivery address: ")
    distance = int(input("Enter the distance (in miles) from the store: "))

    # Check if the item is in stock and get the quantity required
    for item in inventory:
        quantity_required = int(input(f"How many packets of {item['item']} are required? "))
        if quantity_required > item["quantity"]:
            print(f"Sorry, {item['item']} is out of stock.")
        else:
            item["quantity"] -= quantity_required
            total_cost += quantity_required * item["cost_per_item"]
            customer_cart.append({"item": item["item"], "quantity": quantity_required, "cost": quantity_required * item["cost_per_item"]})

    # Calculate the delivery charge
    delivery_charge = calculate_delivery_charge(distance)

    # Display the final bill
    print("\nCustomer Details:")
    print("Name:", customer_name)
    print("Delivery Address:", delivery_address)
    print("Delivery Charge:", delivery_charge)

    print("\nItems Purchased:")
    print("S.No   Items      Quantity   Cost")
    for i, item in enumerate(customer_cart, start=1):
        print(f"{i}\t{item['item']}\t{item['quantity']}\t${item['cost']:.2f}")
    print("Total Cost: ${:.2f}".format(total_cost))
    print("Grand Total: ${:.2f}".format(total_cost + delivery_charge))

    choice = input("Do you want to continue shopping? (yes/no): ")
    if choice.lower() == "no":
        break
```
