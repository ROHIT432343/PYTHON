```python
import pandas as pd
car_data= pd.read_csv('DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv')
```

```python
car_data
```

###### task 1

```python
# Check for missing values
print("Missing Values:")
print(car_data.isnull().sum())

# Drop rows with missing values
df_cleaned = car_data.dropna()

# Display the cleaned-up DataFrame
print("\nCleaned-up DataFrame:")
print(df_cleaned)
```

##### task 2

```python
# Filter the DataFrame for cars from the year 2014 with California as the state
potential_cars = car_data[(car_data['year'] == 2014) & (car_data['state'] == 'ca')]

# Sort the DataFrame by price in ascending order
potential_cars_sorted = potential_cars.sort_values(by='mp')

# Select the top 100 rows
top_100_cars = potential_cars_sorted.head(100)

# Display the details of potential cars
print("Details of Potential Cars:")
print(top_100_cars)
```

###### task 3

```python
# Pivot table to display the top 100 cars based on car manufacturers and prices
cross_table = top_100_cars.pivot_table(index='make', values='mp', aggfunc='min')

# Display the cross table
print("Cross Table of Top 100 Cars:")
print(cross_table)
```

###### task 4

```python
# Filter the top 100 cars DataFrame for Chevrolet cars
chevrolet_cars = top_100_cars[top_100_cars['make'] == 'Chevrolet']

# Calculate the discounted price with a 10% discount
chevrolet_cars['discounted_price'] = chevrolet_cars['mp'] * 0.9

# Display the updated data for Chevrolet cars with original and discounted prices
print("Updated Data for Chevrolet Cars:")
print(chevrolet_cars[['make', 'model', 'year', 'mp', 'discounted_price']])
```

###### task 5

```python
# Filter the DataFrame for cars from the year 2014
cars_2014 = car_data[car_data['year'] == 2014]

# Sort the DataFrame by price in ascending order
lowest_price_cars_2014 = cars_2014.sort_values(by='mp').head(100)

# Calculate the average price based on interior color
average_price_by_color = lowest_price_cars_2014.groupby('interior')['mp'].mean()

# Display the average price based on interior color
print("Average Price of 100 Lowest-Priced Cars from 2014 by Interior Color:")
print(average_price_by_color)
```

```python
# Filter the DataFrame for cars from the year 2010 to 2015
cars_2010_2015 = car_data[(car_data['year'] >= 2010) & (car_data['year'] <= 2015)]

# Create a pivot table to show color-wise availability for each manufacturer
color_availability = cars_2010_2015.pivot_table(index='make', columns='color', values='mp', aggfunc='count', fill_value=0)

# Display the color-wise availability for each manufacturer
print("Color-wise Availability of Cars for Each Manufacturer (2010-2015):")
print(color_availability)
```

###### task 7

```python
# Filter the DataFrame for BMW sedans manufactured between 2010 and 2015 with condition score >= 4.5
bmw_sedans = car_data[(car_data['make'] == 'BMW') & (car_data['body'] == 'Sedan') & (car_data['year'] >= 2010) & (car_data['year'] <= 2015) & (car_data['condition'] >= 4.5)]

# Group by year and model, and find the maximum expected price
max_price_by_year_model = bmw_sedans.groupby(['year', 'model'])['mp'].max()

# Display the details of BMW sedans and the year-wise maximum expected price
print("Details of BMW Sedans with Condition >= 4.5 (2010-2015):")
print(bmw_sedans[['make', 'model', 'year', 'condition', 'mp']])

print("\nYear-wise Maximum Expected Price of BMW Sedans (2010-2015) Grouped by Model:")
print(max_price_by_year_model)
```
