###### task 8

```python
# Step 1: List customers who have property in a rural or semiurban area, applicant income > 4000,
# co-applicant income > 1000, and loan status is 'N'
filtered_customers = loan_data[
    ((loan_data['Property_Area'] == 'Rural') | (loan_data['Property_Area'] == 'Semiurban')) &
    (loan_data['ApplicantIncome'] > 4000) &
    (loan_data['CoapplicantIncome'] > 1000) &
    (loan_data['Loan_Status'] == 'N')
]

# Display information of customers meeting the conditions
print("Customers meeting the conditions:")
print(filtered_customers)
# Count the number of such customers
count_before_update = len(filtered_customers)
print("\nNumber of customers meeting the conditions before update:", count_before_update)

# Step 2: Update the loan status of these customers to 'Y'
loan_data.loc[filtered_customers.index, 'Loan_Status'] = 'Y'

# Step 3: Verify that the count of such customers is now zero
filtered_customers_after_update = loan_data[
    ((loan_data['Property_Area'] == 'Rural') | (loan_data['Property_Area'] == 'Semiurban')) &
    (loan_data['ApplicantIncome'] > 4000) &
    (loan_data['CoapplicantIncome'] > 1000) &
    (loan_data['Loan_Status'] == 'N')
]
# Count the number of such customers after the update
count_after_update = len(filtered_customers_after_update)
print("\nNumber of customers meeting the conditions after update:", count_after_update)

# Display the updated DataFrame
print("\nUpdated DataFrame:")
print(loan_data)
```

```python
# Assuming the DataFrame is already loaded and named loan_data

# Step 1: Calculate the average income of customers living in an urban area
average_urban_income = loan_data[loan_data['Property_Area'] == 'Urban']['ApplicantIncome'].mean()

# Step 2: Filter customers who live in rural or semiurban areas, have 3+ dependents,
# and have income greater than the average income of urban customers
filtered_customers = loan_data[
    ((loan_data['Property_Area'] == 'Rural') | (loan_data['Property_Area'] == 'Semiurban')) &
    (loan_data['Dependents'] == '3+') &
    (loan_data['ApplicantIncome'] > average_urban_income)
]

# Display information of customers meeting the conditions
print("Customers meeting the conditions:")
print(filtered_customers)

# Step 3: Identify an important parameter that the bank may be using to approve the loan
# In this case, it appears that income is a key parameter, as it is being compared to the average income in an urban area.

# Step 4: Display the average applicant income of customers from non-urban areas with loan status 'Y'
average_non_urban_income = loan_data[(loan_data['Property_Area'] != 'Urban') & (loan_data['Loan_Status'] == 'Y')]['ApplicantIncome'].mean()

# Display the average applicant income
print("\nAverage applicant income of customers from non-urban areas with loan status 'Y':", average_non_urban_income)
```
