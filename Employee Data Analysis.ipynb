```python
import pandas as pd 
employee_data= pd.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx')
```

```python
employee_data
```

```python
import matplotlib.pyplot as plt
df = pd.DataFrame(employee_data)

# Summarize the department-wise sum of annual salary
department_salary_sum = df.groupby('Department')['Annual Salary ($)'].sum()

# Plot the bar graph
department_salary_sum.plot(kind='bar', color='skyblue')
plt.xlabel('Department')
plt.ylabel('Total Annual Salary ($)')
plt.title('Total Annual Salary for Employees in Each Department')
plt.show()
```

```python
# Summarize the department and gender-wise count of employees
department_gender_count = df.groupby(['Department', 'Gender']).size().unstack(fill_value=0)

# Create a DataFrame with department name and representative male and female count
department_gender_df = pd.DataFrame({
    'Department': department_gender_count.index,
    'Male_Count': department_gender_count['Male'],
    'Female_Count': department_gender_count['Female']
})

# Display the resulting DataFrame
print("Department-wise Total Number of Male and Female Employees:")
print(department_gender_df)
# Bar chart for female count
plt.bar(department_gender_df['Department'], department_gender_df['Female_Count'], bottom=department_gender_df['Male_Count'], label='Female', color='pink')

# Add labels and title
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.title('Department-wise Total Number of Male and Female Employees')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
# Filter the data for employees with zero work experience in the IT department
filtered_data = employee_data[(employee_data['Work_Experience'] == 0) & (employee_data['Department'] == 'IT ')]
# Calculate the average salary for the filtered data
average_salary = filtered_data['Annual Salary ($)'].mean()

# Display the average salary for a new employee in the IT department
print("Estimated average salary for a new employee in the IT department:", average_salary)
```

```python
# Summarize data by department
department_salary_sum = df.groupby('Department')['Annual Salary ($)'].sum()

# Plot a pie chart for department-wise ratio of the sum of annual salary
plt.figure(figsize=(8, 8))
plt.pie(department_salary_sum, labels=department_salary_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('Department-wise Ratio of Sum of Annual Salary')
plt.show()
```

###### task 5

```python
# Assuming the DataFrame is already loaded and named df

# Define age group bins
bins = [25, 30, 35, 40, 45, 50]

# Plot a histogram for the distribution of employee count by age group
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=bins, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employee Count by Age Group')
plt.xticks(bins)
plt.grid(axis='y', linestyle='--')
plt.show()
```

###### task 6

```python
# Plot a scatter plot for salary trend based on employee experience
plt.figure(figsize=(10, 6))
plt.scatter(df['Work_Experience'], df['Annual Salary ($)'], color='blue')
plt.xlabel('Work Experience (years)')
plt.ylabel('Annual Salary ($)')
plt.title('Salary Trend Based on Employee Experience')
plt.grid(True, linestyle='--')
plt.show()
```
